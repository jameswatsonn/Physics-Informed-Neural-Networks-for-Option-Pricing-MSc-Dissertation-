# 1D Black-Scholes (European Call Option)

```python
import sys
import math
import logging
import matplotlib.pyplot as plt
from absl import app
from main import main
from datetime import datetime

# Black-Scholes closed-form solution
def black_scholes_call(S, K, T, r, sigma):
    d1 = (math.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    N1 = (1.0 + math.erf(d1 / math.sqrt(2.0))) / 2.0
    N2 = (1.0 + math.erf(d2 / math.sqrt(2.0))) / 2.0
    call_price = S * N1 - K * math.exp(-r * T) * N2
    return call_price

# Parameters
S0 = 100.0 # Initial stock price (at S_max)
K = 50.0  # Strike price
T = 1.0  # Time to maturity
r = 0.05 # Risk-free rate (mu in config)
sigma = 0.25  # Volatility

# Calculate true option price at t=0 for S0=100
true_price = black_scholes_call(S0, K, T, r, sigma)
print(f"True European Call Option Price: {true_price:.4f}")

# Log handler to extract Y0 (netwrok prediction) and compute error
class ErrorLogHandler(logging.Handler):
    def __init__(self, true_price):
        super().__init__()
        self.true_price = true_price
        self.error_data = [] 
        
    def emit(self, record):
        msg = record.getMessage()
        if "step:" in msg and "Y0:" in msg:
            try:
                # Extract step number
                step_str = msg.split("step:")[1].split(",")[0].strip()
                step = int(step_str)
                
                # Extract Y0 value
                y0_str = msg.split("Y0:")[1].split(",")[0].strip()
                y0_val = float(y0_str)
                
                # Compute L2 relative error
                if self.true_price != 0:
                    error = abs(y0_val - self.true_price) / self.true_price
                else:
                    error = float('nan')
                
                # Store error data for plotting
                self.error_data.append((step, error))
                
                # Print error
                print(f"Step {step}: L2 Relative Error: {error:.4e}")
            except (ValueError, IndexError, TypeError) as e:
                print(f"Error parsing log: {e}")
                pass  # Skip if extraction fails

# Plot L2-relative error progression over training steps
def plot_error_progression(error_data, true_price):
    if not error_data:
        print("No error data collected for plotting")
        return
        
    steps, errors = zip(*error_data)
    
    plt.figure(figsize=(10, 6))
    plt.plot(steps, errors, 'b-o', linewidth=2, markersize=6)
    plt.xlabel('Training Step', fontsize=12)
    plt.ylabel('L2 Relative Error', fontsize=12)
    plt.title(f'Error Progression | True Price: {true_price:.4f}', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Axis limits
    if min(errors) > 0:
        plt.ylim(bottom=0, top=max(errors)*1.1)
    
    plt.tight_layout()
    
    # Save plot to file
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"error_progression_{timestamp}.png"
    plt.savefig(filename, dpi=300)
    print(f"Saved error progression plot to {filename}")
    
    plt.show()

def run_main():
    # Configure logging handler
    root_logger = logging.getLogger()
    error_handler = ErrorLogHandler(true_price)
    error_handler.setLevel(logging.INFO)
    root_logger.addHandler(error_handler)
    
    try:
        # Override sys.argv with updated config path
        sys.argv = ['script_name.py', '--config_path=configs/1D_BS_European_Call.json', '--exp_name=BlackScholes'] # Load configuration file 
        app.run(main)
    except SystemExit:
        pass  # Handle termination
    finally:
        # Clean up logging handler
        root_logger.removeHandler(error_handler)
        
        # Plot error progression after training completes
        if error_handler.error_data:
            plot_error_progression(error_handler.error_data, true_price)
        else:
            print("No error data available for plotting")

if __name__ == '__main__':
    run_main()
```

```python
import sys
import math
import logging
import json
import numpy as np
import tensorflow as tf
import time
import matplotlib.pyplot as plt
from absl import app
from main import main
from datetime import datetime
import QuantLib as ql  # Import QuantLib

# Compute European call price under Heston dynamics using QuantLib's analytic MC simulation engine - From QuantLib GitHub 
def heston_analytic_price(S0, v0, T, r, kappa, theta, sigma, rho, strike):
    # Set arbitrary evaluation date
    evaluation_date = ql.Date(1, ql.January, 2000)
    ql.Settings.instance().evaluationDate = evaluation_date
    
    # Calculate exercise date (T years from evaluation date)
    exercise_date = evaluation_date + int(round(T * 365.25))
    
    # Option setup  - directly from QuantLib GitHub 
    exercise = ql.EuropeanExercise(exercise_date) # European-style - payoff only at maturity 
    payoff = ql.PlainVanillaPayoff(ql.Option.Call, strike) # payoff function max(S-K, 0)
    option = ql.VanillaOption(payoff, exercise) # Create European call option 
    
    # Setup market data - directly from QuantLib GitHub 
    underlying = ql.SimpleQuote(S0) # Current asset price 
    risk_free_curve = ql.FlatForward(evaluation_date, r, ql.Actual365Fixed()) # risk-free rate term 
    dividend_curve = ql.FlatForward(evaluation_date, 0.0, ql.Actual365Fixed()) # No dividends
    
    # Build Heston process and model - directly from QuantLib GitHub 
    heston_process = ql.HestonProcess( ql.YieldTermStructureHandle(risk_free_curve), ql.YieldTermStructureHandle(dividend_curve), ql.QuoteHandle(underlying),
        v0, # initial variance
        kappa,
        theta, # long-term variance
        sigma,  # volatility of volatility
        rho)
    heston_model = ql.HestonModel(heston_process)
    
    # Price using QuantLib engine
    option.setPricingEngine(ql.AnalyticHestonEngine(heston_model))
    return option.NPV()

# Log handler to extract Y0 (network prediction) and compute error
class ErrorLogHandler(logging.Handler):
    def __init__(self, true_price):
        super().__init__()
        self.true_price = true_price
        self.error_data = [] 
        
    def emit(self, record):
        msg = record.getMessage()
        if "step:" in msg and "Y0:" in msg:
            try:
                # Extract step number
                step_str = msg.split("step:")[1].split(",")[0].strip()
                step = int(step_str)
                
                # Extract Y0 value
                y0_str = msg.split("Y0:")[1].split(",")[0].strip()
                y0_val = float(y0_str)
                
                # Compute L2 relative error
                if self.true_price != 0:
                    error = abs(y0_val - self.true_price) / self.true_price
                else:
                    error = float('nan')
                
                # Store error data for plotting
                self.error_data.append((step, error))
                
                # Print error 
                print(f"Step {step}: Y0={y0_val:.4f}, True={self.true_price:.4f}, Error={error:.4e}")
            except (ValueError, IndexError, TypeError) as e:
                print(f"Error parsing log: {e}")
                pass  # Skip if extraction fails

# Plot L2-relative error progression over training step
def plot_error_progression(error_data, true_price):
    if not error_data:
        print("No error data collected for plotting")
        return
        
    steps, errors = zip(*error_data)
    
    plt.figure(figsize=(10, 6))
    plt.plot(steps, errors, 'b-o', linewidth=2, markersize=6)
    plt.xlabel('Training Step', fontsize=12)
    plt.ylabel('L2 Relative Error', fontsize=12)
    plt.title(f'Heston Model Error Progression | True Price: {true_price:.4f}', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Axis limits
    if min(errors) > 0:
        plt.ylim(bottom=0, top=max(errors)*1.1)
    
    plt.tight_layout()
    
    # Save plot to file
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"heston_error_progression_{timestamp}.png"
    plt.savefig(filename, dpi=300)
    print(f"Saved error progression plot to {filename}")
    
    plt.show()


def run_main():
    # Load Heston configuration file 
    config_path = 'configs/1D_Heston_European_Call.json'
    with open(config_path) as f:
        config = json.load(f)
    
    # Extract model parameters from config file 
    params = config['eqn_config']['model_params']
    S0 = params['initial_stock']
    v0 = params['initial_vol']  # initial variance
    r = params['r']
    kappa = params['kappa']
    theta = params['theta']  # long-term variance
    epsilon = params['epsilon'] # vol of vol
    rho = params['rho']
    strike = params['strike']
    T = config['eqn_config']['total_time']
    
    # Calculate true option price using QuantLib
    print("Computing true price via QuantLib Analytic Heston Engine...")
    start_time = time.time()
    true_price = heston_analytic_price(S0, v0, T, r, kappa, theta, epsilon, rho, strike)
    elapsed = time.time() - start_time
    print(f"QuantLib Analytic Price: {true_price:.6f} (took {elapsed:.2f} seconds)")

    # Configure logging handler
    root_logger = logging.getLogger()
    error_handler = ErrorLogHandler(true_price)
    error_handler.setLevel(logging.INFO)
    root_logger.addHandler(error_handler)
    
    try:
        # Override sys.argv with Heston config path
        sys.argv = ['script_name.py', f'--config_path={config_path}', '--exp_name=Heston']
        app.run(main)
    except SystemExit:
        pass
    finally:
        # Clean up and plot
        root_logger.removeHandler(error_handler)
        if error_handler.error_data:
            plot_error_progression(error_handler.error_data, true_price)
        else:
            print("No error data available for plotting")

if __name__ == '__main__':
    run_main()
```
